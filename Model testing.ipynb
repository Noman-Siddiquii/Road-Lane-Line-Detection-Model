```python
import numpy as np
import cv2

def plot_predicted_image(model, input_image):
  """
  Plots the output image generated by the image-to-image CNN model, in black and white.

  Args:
      model: The trained Keras image-to-image CNN model.
      input_image: The input image as a NumPy array.

  Returns:
      The predicted image as a grayscale NumPy array.
  """

  # Preprocess the input image if necessary (e.g., normalization, resizing)
  # Assuming your model expects preprocessed input, uncomment these lines:
  # input_image = preprocess_input(input_image)  # Replace with your preprocessing function

  # Make prediction
  predicted_image = model.predict(np.expand_dims(input_image, axis=0))[0]

  # Reshape the predicted image to match the expected output shape

  # Convert the predicted image to grayscale (weighted average)
  grayscale_image = np.average(predicted_image, axis=-1)

  # Return the grayscale image
  return   grayscale_image

# Example usage (assuming your model is loaded as 'model' and input image is 'img')
predicted_image = plot_predicted_image(model=loaded_model, input_image=img)

# Display the predicted image in grayscale using OpenCV
cv2.imshow("Predicted Image (Grayscale)", predicted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("Predicted image displayed in grayscale.")
```
